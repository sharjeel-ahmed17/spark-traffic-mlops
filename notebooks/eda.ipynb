```python
# To rename the column
df = df.rename({"ID":"Vehicle ID,"},axis=1)
```

```python
# To find and sum the missing values
df.isna().sum()
```

```python
# To find out the mean of the column
mean = df["Vehicles"].mean().round()
mean
```

```python
# To find and replace the missing values with the mean
df["Vehicles"] = df["Vehicles"].fillna(mean)
```
